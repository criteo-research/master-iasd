```python
import pandas
import sys
from pandas.api.types import CategoricalDtype

def print_stats(df):
    nb_bytes = sys.getsizeof(df)
    nb_rows = len(df.index)
    print(f"Total MB {nb_bytes / 1000000}")
    print(f"Nb Rows : {nb_rows}")
    print(f"Row Bytes : {nb_bytes/nb_rows}")

big_file = 'big_dataset.csv'
```

```python
df = pandas.read_csv(big_file, usecols=['temperature'], nrows=1000)
df["temperature"]  = df["temperature"].astype("float32")
col_types = {}
col_types["temperature"] = df.dtypes["temperature"]
df_final = pandas.read_csv(big_file, usecols=["temperature"], dtype=col_types, nrows=1000)
```

```python
print_stats(df_final)
```

```python
col_types = {}
col_types["Acorn"] = CategoricalDtype(categories=["ACORN-A"])
df_final = pandas.read_csv(big_file, usecols=["Acorn"], dtype=col_types, nrows=10000)
print_stats(df_final)
```
